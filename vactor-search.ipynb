```python
documentLoder = TextLoader("tagged_description.txt").load()
text_Splitter = CharacterTextSplitter(separator="\n",chunk_overlap=0,chunk_size=0)
documents = text_Splitter.split_documents(documentLoder)
```

```python
documents[0]
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
db_books = Chroma.from_documents(documents, embedding=embedding)
```

```python

query = "What is the book about?"
results = db_books.similarity_search(query, k=5)
for result in results:
    print(result.page_content)
    print("\n---\n")
    
```

```python
def retrieve_semantic_recommendations(
      query: str,
      # number of recommendations we want
      top_k: int = 10
) -> pd.DataFrame:
    # Get similar documents from vector DB
    recs = db_books.similarity_search(query, k=50)

    isbns = []
    for rec in recs:  # Iterate through recs directly
        try:
            isbn_str = rec.page_content.split()[0].strip()  # Get first word
            isbn_int = int(isbn_str)
            isbns.append(isbn_int)  # Append to list properly
        except ValueError:
            # Skip if ISBN can't be converted to integer
            continue

    # Filter books and then limit results
    recommended_books = book[book["isbn13"].isin(isbns)].head(top_k)
    return recommended_books
```

```python
retrieve_semantic_recommendations("a book to teach children about nature ")
```

```python
def retrive_semetric_number(
        query : str ,

        top_key: int=10
) -> pd.DataFrame :

    search_sematric_from_documents= db_books.similarity_search(query,k=50)

    list_documents=[]

    for i in search_sematric_from_documents:
        try:
            isbn_str= i.page_content.split()[0].strip()
            isbn_int=int(isbn_str)
            list_documents.append(isbn_int)

        except ValueError:
            continue

    recommended_books = book[book["isbn13"].isin(list_documents)].head(top_key)
    return recommended_books



```

```python
retrive_semetric_number("a book ")
```
